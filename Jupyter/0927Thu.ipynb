### Bubble Sort

```python
num = [5,4,1,3,9,2,15,7,6,8,10]

for i in range(0, len(num)):
    print("Step", i, "-----------")
    for j in range(0, len(num) - i - 1):
        if num[j] > num[j+1]:
            temp = num[j]
            num[j] = num[j+1]
            num[j+1] = temp
            print(num)
print("Sort Done")
print(num)
```

```python
num = int(input("input num : "))

for i in range(2, num):
    if num%i == 0:
        break

if i == num-1:
    print(num, "is prime number")
else:
    print(num, "is not prime number")
```
